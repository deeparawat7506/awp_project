```python
#Second method to find the outlier : Inter Quantile Range (IQR)
# find first quantile Q1 : quantile() inbuilt method 
Q1=df['Price'].quantile(.25) #.25 means 25%
# find Third  quantile Q3
Q3=df['Price'].quantile(.75) #.75 means 75%
#Find the Inter Quantile range IQR
IQR=Q3-Q1
#find Lower Whisker
LW=Q1-1.5*IQR
#find upper whisker
UW=Q3+1.5*IQR
print("First Quantile : ",Q1)
print("Third Quantile : ",Q3)
print("Inter Quantile range IQR",IQR)
print("Lower Whisker : ",LW)
print("upper whisker : ",UW)
```

```python
#Find outlier 
a=df[(LW>df['Price'])|(UW<df['Price'])]
a
```
